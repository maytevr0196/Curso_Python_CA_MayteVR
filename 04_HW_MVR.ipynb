-  Plot histogram of DIC and TA

```python
import pandas as pd
 
CO2Data = pd.read_csv("../Curso_Python_Mayte/data/Terminos_lagoon_TA_DIC_2023_RawData.csv")

print(CO2Data.info())
```

```python
import matplotlib.pyplot as plt

# Select data
DIC = CO2Data["dic_micromol_kg"]
TA = CO2Data["ta_micromol_kg"]

plt.hist(DIC, bins=20, density=True, edgecolor="black", alpha=0.85, color='seagreen')
plt.xlabel("DIC (mol kg$^{-1}$)")
plt.ylabel("Probability")
plt.title("Histogram of DIC (density)")
plt.show()

plt.hist(TA, bins=20, density=True, edgecolor="black", alpha=0.85, color='orchid')
plt.xlabel("TA (mol kg$^{-1}$)")
plt.ylabel("Probability")
plt.title("Histogram of TA (density)")
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(10,4), constrained_layout=True)

axes[0].hist(DIC, bins=20, density=True, edgecolor="black", alpha=0.85, color='seagreen')
axes[0].set_title("DIC")
axes[0].set_xlabel("mol kg$^{-1}$")
axes[0].set_ylabel("Probability")

axes[1].hist(TA, bins=20, density=True, edgecolor="black", alpha=0.85, color='orchid')
axes[1].set_title("TA")
axes[1].set_xlabel("mol kg$^{-1}$")

plt.show()
```
